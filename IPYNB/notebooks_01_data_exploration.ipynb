```python
import pandas as pd

df = pd.read_csv('/home/events_sample.csv')

# Basic Info
print(df.shape)
print(df.head())

# Check event distribution
print(df['event_type'].value_counts())

# Unique users
print("Unique Users:", df['user_id'].nunique())

# Convert time
df['event_time'] = pd.to_datetime(df['event_time'])

# Events per user
events_per_user = df.groupby('user_id').size().mean()
print("Avg events per user:", events_per_user)
```
